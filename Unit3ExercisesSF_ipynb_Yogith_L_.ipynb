## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**: Describe what processes "generate" each of the previous discrete distributions models. Also report their support. Here's a spicier way to phrase the question: describe these distributions' maximum entropy conditions in english. Hint: Make sure to read any of the: preliZ documentation, relevant wikipedia articles, or wolfram alpha.

**Maximum entropy conditions:** Highest entropy is typically reached when probability is uniform.


**Bernoulli**

Process: Models binary outcomes of a single trial

Support: 0,1

Maximum Entropy Condition: Maximum entropy occurs at p=.5 because this is where the outcomes for trials are the most uncertain/random.  

**Binomial**

Process: Models the number of sucesses in n trials

Support: 0,n

Maximum Entropy Condition: Similar to the bernoulli distribution, entropy is maximized at p=.5 since sucesss and failure are both equally probable.

**Poisson**:

Process: Models the number of events occurring in a fixed period of time when the times are independent

Support: $[0,∞]$

Maximum Entropy Condition: Entropy is maximized when the parameter, mu, is increased since this causes the graph to become more uniform, making the probability of each event more similar.

**Geometric**

Process: The probability that the first success in a series of Bernoulli trials occurrs in the xth trial given the probability of success

Support: $x \in \mathbb{N}_{>0}$ (the set of values the random variable can take)

Maximum Entropy Condition: Entropy is maximized when p approaches 0. This is evident in the graph as it becomes more unifrom when probability nears 0.

**HyperGeometric**

Process: The probability of x successses in n fixed bernoulli trials taken w/o replacement from a population of N object containing k successess

Support: $x \in \left[\max(0, n - N + k), \min(k, n)\right]$ (the set of integers k can take)

Maximum Entropy Condition: Occurs when k/n=.5. In otherwords, the quotient of the fixed number of sucessess and fixed population sample need to equal .5 to maximize entropy.

**Categorical**

Process: Models the probability of each outcome in a set of categories.

Support: $x \in \{0, 1, \ldots, |p|-1\}$, (the set of categories it can take)

Maximum Entropy Condition: All the values of p should be equal so their can be a unifrom distribution across categories.
